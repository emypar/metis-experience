```python
df = pd.read_csv("station_data.csv", index_col=0, header=0)

df.drop(labels='DATE_TIME', axis=1, inplace=True, errors='ignore')

df_tot = (
    df.groupby(['DAYS', 'STATION_LINE'])
    .sum()
    .reset_index()
    .sort_values(by='TOTAL_TRAFFIC', ascending=False)
    .set_index('DAYS', drop=True)
)
```
