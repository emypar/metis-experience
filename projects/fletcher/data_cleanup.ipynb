```python
import sys
import os
import time
```

```python
import pandas as pd
```

```python
# Load the train set and remove NaN's and duplicates:
train_file = './data/quora/train.csv.zip'
df = pd.read_csv(train_file, header=0)
df.shape
```

```python
clean_df = df[['question1', 'question2', 'is_duplicate']].dropna()
clean_df.drop_duplicates(subset=['question1', 'question2'], inplace=True)
clean_df.shape
```

```python
%%time
# Keep only English questions:
from langdetect import detect as lang_detect
from langdetect.lang_detect_exception import LangDetectException

n = 0
start_time = time.time()

def q12_en(row):
    global n
    n += 1
    if not (n % 1000):
        print("dtime={:+12.06f} s, n={}".format(
            time.time() - start_time,
            n
        ))
    try:
        both_en = (
            lang_detect(row['question1']) == 'en'
            and lang_detect(row['question2']) == 'en'
        )
    except (LangDetectException) as e:
        both_en = False
    return both_en

en_filter = clean_df.apply(q12_en, axis=1)
clean_df = clean_df[en_filter]
```

```python
clean_df.shape
```

```python
clean_file = './data/quora/clean.csv.gz'
clean_df.to_csv(clean_file, compression='gzip', index=False)
```
